## üë®üèª‚Äçüíª Factorial  example

- general formual for any number n
    - n! = n * (n-1)!
    - 0! = 1
    - 1! = 1
- recursion tree, remember calculation is happening from top to up
    

*Code snippet 2 -- find factorial of a number*

```python
def fact(n):
    if n <= 1:
        return 1
    else:
        return n * fact(n-1)

print(fact(3))
```

## üë®üèª‚Äçüíª Fibonacci example

- recursion tree
    

*Code snippet 3 -- find fibonacci factorial*

```python
def fibo(n):
    if n <= 1:
        return n
    return fibo(n-1)+fibo(n-2)

print(fibo(7))
```
